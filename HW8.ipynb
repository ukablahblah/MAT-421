```python
# Numerical Integration: Riemann's Integral and Trapezoid Rule
# Numerical integration is a method to approximate the definite integral of a function
# when an analytical solution is difficult or impossible to obtain.

# Importing necessary libraries
import numpy as np
import matplotlib.pyplot as plt

# Riemann's Integral
# Explanation:
# Riemann's Integral approximates the area under a curve by dividing it into small rectangles.
# The sum of the areas of these rectangles gives an approximation of the integral.

# Function to calculate Riemann's Integral
def riemann_integral(f, a, b, n, method='left'):
    """
    Approximates the integral of a function f(x) using Riemann's Sum.

    Parameters:
    f: function - The function to integrate
    a: float - The start of the interval
    b: float - The end of the interval
    n: int - The number of subintervals
    method: str - 'left', 'right', or 'midpoint' for the Riemann sum method

    Returns:
    float - The approximate integral
    """
    x = np.linspace(a, b, n+1)  # Create n+1 points between a and b
    dx = (b - a) / n  # Width of each subinterval

    if method == 'left':
        x_left = x[:-1]  # Left endpoints
        return np.sum(f(x_left) * dx)
    elif method == 'right':
        x_right = x[1:]  # Right endpoints
        return np.sum(f(x_right) * dx)
    elif method == 'midpoint':
        x_mid = (x[:-1] + x[1:]) / 2  # Midpoints
        return np.sum(f(x_mid) * dx)
    else:
        raise ValueError("Method must be 'left', 'right', or 'midpoint'.")

# Example: Approximate the integral of f(x) = x^2 from 0 to 1
f = lambda x: x**2
a, b = 0, 1
n = 10  # Number of subintervals

# Calculate Riemann sums
left_sum = riemann_integral(f, a, b, n, method='left')
right_sum = riemann_integral(f, a, b, n, method='right')
midpoint_sum = riemann_integral(f, a, b, n, method='midpoint')

# Print results
print("Riemann's Integral Approximation:")
print(f"Left Riemann Sum: {left_sum}")
print(f"Right Riemann Sum: {right_sum}")
print(f"Midpoint Riemann Sum: {midpoint_sum}")

# Visualization of Riemann sums
x = np.linspace(a, b, 100)
plt.plot(x, f(x), label='f(x) = x^2', color='blue')
x_rect = np.linspace(a, b, n+1)
dx = (b - a) / n

# Left Riemann Sum
x_left = x_rect[:-1]
plt.bar(x_left, f(x_left), width=dx, align='edge', alpha=0.5, label='Left Riemann Sum')

# Right Riemann Sum
x_right = x_rect[1:]
plt.bar(x_right, f(x_right), width=-dx, align='edge', alpha=0.5, label='Right Riemann Sum')

plt.title("Riemann's Integral Approximation")
plt.xlabel("x")
plt.ylabel("f(x)")
plt.legend()
plt.show()

```

```python
# Trapezoid Rule
# Explanation:
# The Trapezoid Rule approximates the area under a curve by dividing it into trapezoids.
# The sum of the areas of these trapezoids gives an approximation of the integral.

# Function to calculate the Trapezoid Rule
def trapezoid_rule(f, a, b, n):
    """
    Approximates the integral of a function f(x) using the Trapezoid Rule.

    Parameters:
    f: function - The function to integrate
    a: float - The start of the interval
    b: float - The end of the interval
    n: int - The number of subintervals

    Returns:
    float - The approximate integral
    """
    x = np.linspace(a, b, n+1)  # Create n+1 points between a and b
    y = f(x)  # Evaluate the function at each point
    dx = (b - a) / n  # Width of each subinterval
    return (dx / 2) * np.sum(y[:-1] + y[1:])  # Trapezoid Rule formula

# Example: Approximate the integral of f(x) = x^2 from 0 to 1 using the Trapezoid Rule
trapezoid_sum = trapezoid_rule(f, a, b, n)

# Print result
print("\nTrapezoid Rule Approximation:")
print(f"Trapezoid Rule Sum: {trapezoid_sum}")

# Visualization of the Trapezoid Rule
plt.plot(x, f(x), label='f(x) = x^2', color='blue')
x_trap = np.linspace(a, b, n+1)
y_trap = f(x_trap)

for i in range(n):
    plt.fill([x_trap[i], x_trap[i], x_trap[i+1], x_trap[i+1]],
             [0, y_trap[i], y_trap[i+1], 0], 'orange', edgecolor='black', alpha=0.5)

plt.title("Trapezoid Rule Approximation")
plt.xlabel("x")
plt.ylabel("f(x)")
plt.legend()
plt.show()

```
