```python
import numpy as np
import matplotlib.pyplot as plt

def numerical_derivative(f, x, h=1e-5, method='central'):
    """

    Parameters:
    f (function): Function to differentiate
    x (float or array): Point(s) at which to compute the derivative
    h (float): Step size
    method (str): 'forward', 'backward', or 'central'

    Returns:
    float or array: Approximated derivative
    """
    if method == 'forward':
        return (f(x + h) - f(x)) / h
    elif method == 'backward':
        return (f(x) - f(x - h)) / h
    elif method == 'central':
        return (f(x + h) - f(x - h)) / (2 * h)
    else:
        raise ValueError("Invalid method. Choose 'forward', 'backward', or 'central'.")

# Example function: f(x) = x^2
f = lambda x: x**2
x_val = 2.0

# Compute derivatives using different methods
forward_diff = numerical_derivative(f, x_val, method='forward')
backward_diff = numerical_derivative(f, x_val, method='backward')
central_diff = numerical_derivative(f, x_val, method='central')

print(f"Forward Difference Approximation: {forward_diff}")
print(f"Backward Difference Approximation: {backward_diff}")
print(f"Central Difference Approximation: {central_diff}")

# Higher-order derivatives approximation
def second_derivative(f, x, h=1e-5):
    """Compute second derivative using central difference."""
    return (f(x + h) - 2 * f(x) + f(x - h)) / (h**2)

second_diff = second_derivative(f, x_val)
print(f"Second Derivative Approximation: {second_diff}")

# Visualization of finite differences
x = np.linspace(-2, 2, 100)
y = f(x)
deriv_y = [numerical_derivative(f, xi) for xi in x]

plt.figure(figsize=(8,5))
plt.plot(x, y, label='f(x) = x^2')
plt.plot(x, deriv_y, label="f'(x) Approximation", linestyle='dashed')
plt.legend()
plt.xlabel('x')
plt.ylabel('y')
plt.title('Numerical Differentiation')
plt.grid()
plt.show()
```
