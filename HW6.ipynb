```python
# Continuity and Differentiation, Taylor Theorem, and Optimization with Gradient Descent

# Importing necessary libraries
import numpy as np
import matplotlib.pyplot as plt

# Section 1: Continuity and Differentiation

#Continuity: A function is continuous if there are no breaks, jumps, or holes in its graph
#Differentiation: Differentiation measures the rate of change of a function.
#The derivative of a function f(x) is defined as the limit of the difference quotient as h approaches 0

# Example: Differentiating a simple function
def f(x):
    return x**2 + 3*x + 2  # Define the function

def f_prime(x, h=1e-5):
    return (f(x + h) - f(x - h)) / (2 * h)  # Numerical derivative using central difference

# Generate x values
x_vals = np.linspace(-10, 10, 100)
f_vals = f(x_vals)
f_prime_vals = f_prime(x_vals)

# Plotting the function and its derivative
plt.figure(figsize=(10, 6))
plt.plot(x_vals, f_vals, label="f(x) = x^2 + 3x + 2", color='blue')
plt.plot(x_vals, f_prime_vals, label="f'(x) = 2x + 3 (Numerical Derivative)", color='red', linestyle='--')
plt.axhline(0, color='black', linewidth=0.5, linestyle='--')
plt.title("Function and its Derivative")
plt.xlabel("x")
plt.ylabel("y")
plt.legend()
plt.grid()
plt.show()

```

```python
# Section 2: Taylor Theorem
#Taylor Theorem: The Taylor series approximates a function as a polynomial around a point a point
#The nth-degree Taylor polynomial is given by
#T_n(x) = f(a) + f'(a)(x-a) + f''(a)(x-a)^2/2! + ... + f^n(a)(x-a)^n/n!

# Example: Taylor series approximation for e^x around x=0
def taylor_exp(x, n):
    """Compute the Taylor series approximation of e^x around x=0."""
    result = 0
    for i in range(n + 1):
        result += (x**i) / np.math.factorial(i)
    return result

# Generate x values
x_vals = np.linspace(-2, 2, 100)
true_vals = np.exp(x_vals)  # True values of e^x
taylor_approx_2 = taylor_exp(x_vals, 2)  # 2nd-degree Taylor approximation
taylor_approx_5 = taylor_exp(x_vals, 5)  # 5th-degree Taylor approximation

# Plotting the true function and Taylor approximations
plt.figure(figsize=(10, 6))
plt.plot(x_vals, true_vals, label="e^x (True Function)", color='blue')
plt.plot(x_vals, taylor_approx_2, label="Taylor Approximation (n=2)", color='green', linestyle='--')
plt.plot(x_vals, taylor_approx_5, label="Taylor Approximation (n=5)", color='red', linestyle='--')
plt.axhline(0, color='black', linewidth=0.5, linestyle='--')
plt.title("Taylor Series Approximation of e^x")
plt.xlabel("x")
plt.ylabel("y")
plt.legend()
plt.grid()
plt.show()

```

```python
# Section 3: Optimization with Gradient Descent


# Gradient descent is an optimization algorithm used to minimize a function
#It works by iteratively moving in the direction of the negative gradient of the function

# Example: Minimizing a quadratic function f(x) = x^2 + 4x + 4
def quadratic(x):
    return x**2 + 4*x + 4  # Define the quadratic function

def quadratic_grad(x):
    return 2*x + 4  # Gradient of the quadratic function

# Gradient descent implementation
def gradient_descent(grad, lr=0.1, epochs=50, start=10):
    x = start  # Starting point
    history = [x]  # To store the path of x
    for _ in range(epochs):
        x -= lr * grad(x)  # Update x using the gradient
        history.append(x)
    return history

# Run gradient descent
learning_rate = 0.1
epochs = 50
start_point = 10
path = gradient_descent(quadratic_grad, lr=learning_rate, epochs=epochs, start=start_point)

# Plotting the function and gradient descent path
x_vals = np.linspace(-10, 10, 100)
f_vals = quadratic(x_vals)

plt.figure(figsize=(10, 6))
plt.plot(x_vals, f_vals, label="f(x) = x^2 + 4x + 4", color='blue')
plt.scatter(path, quadratic(np.array(path)), color='red', label="Gradient Descent Path")
plt.title("Gradient Descent on f(x) = x^2 + 4x + 4")
plt.xlabel("x")
plt.ylabel("f(x)")
plt.legend()
plt.grid()
plt.show()

print("Gradient Descent completed. Final value of x:", path[-1])
```
