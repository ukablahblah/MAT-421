```python
# Linear Algebra and Linear Regression Demonstration

# Importing necessary libraries
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error

```

```python
# Section 1: Linear Algebra Concepts
# -----------------------------------

# 1. Linear Space
# A linear space (or vector space) is a collection of vectors that can be added together and multiplied by scalars.
# Example: Creating a linear space using NumPy
print("Linear Space Example:")
vector1 = np.array([1, 2, 3])
vector2 = np.array([4, 5, 6])
scalar = 2

# Vector addition
vector_sum = vector1 + vector2
print(f"Vector Addition: {vector1} + {vector2} = {vector_sum}")

# Scalar multiplication
scaled_vector = scalar * vector1
print(f"Scalar Multiplication: {scalar} * {vector1} = {scaled_vector}")

```

```python
# 2. Orthogonality
# Two vectors are orthogonal if their dot product is zero.
print("\nOrthogonality Example:")
vector3 = np.array([1, 0])
vector4 = np.array([0, 1])

dot_product = np.dot(vector3, vector4)
print(f"Dot Product of {vector3} and {vector4}: {dot_product}")
if dot_product == 0:
    print("The vectors are orthogonal.")
else:
    print("The vectors are not orthogonal.")

```

```python
# 3. Eigenvalues and Eigenvectors
# Eigenvalues and eigenvectors are fundamental in linear transformations.
print("\nEigenvalues and Eigenvectors Example:")
matrix = np.array([[4, -2],
                   [1, 1]])

eigenvalues, eigenvectors = np.linalg.eig(matrix)
print(f"Matrix:\n{matrix}")
print(f"Eigenvalues: {eigenvalues}")
print(f"Eigenvectors:\n{eigenvectors}")

```

```python
# Section 2: Linear Regression
# ----------------------------

# Linear regression is a method to model the relationship between a dependent variable and one or more independent variables.

# Example: Simple Linear Regression
print("\nLinear Regression Example:")

# Generating synthetic data
np.random.seed(42)
X = 2 * np.random.rand(100, 1)  # Independent variable
y = 4 + 3 * X + np.random.randn(100, 1)  # Dependent variable with some noise

# Visualizing the data
plt.scatter(X, y, color="blue", label="Data points")
plt.title("Scatter Plot of Data")
plt.xlabel("X")
plt.ylabel("y")
plt.legend()
plt.show()

# Fitting a linear regression model
model = LinearRegression()
model.fit(X, y)

# Model parameters
print(f"Intercept: {model.intercept_[0]}")
print(f"Coefficient: {model.coef_[0][0]}")

# Making predictions
X_new = np.array([[0], [2]])
y_pred = model.predict(X_new)

# Visualizing the regression line
plt.scatter(X, y, color="blue", label="Data points")
plt.plot(X_new, y_pred, color="red", label="Regression line")
plt.title("Linear Regression Model")
plt.xlabel("X")
plt.ylabel("y")
plt.legend()
plt.show()

# Evaluating the model
y_pred_all = model.predict(X)
mse = mean_squared_error(y, y_pred_all)
print(f"Mean Squared Error: {mse}")

```
