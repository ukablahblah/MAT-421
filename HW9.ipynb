```python
# Simpson's Rule Explanation and Implementation

# Importing necessary libraries
import numpy as np
import matplotlib.pyplot as plt

# Introduction to Simpson's Rule
"""
Simpson's Rule is a numerical method for approximating the definite integral of a function.
It is particularly useful when the function is difficult or impossible to integrate analytically.

The formula for Simpson's Rule is:
    ∫(a to b) f(x) dx ≈ (b - a) / 6 * [f(a) + 4f((a + b) / 2) + f(b)]

This method assumes the function can be approximated by a parabola over the interval [a, b].
"""

# Implementing Simpson's Rule
def simpsons_rule(f, a, b):
    """
    Approximates the integral of f(x) from a to b using Simpson's Rule.

    Parameters:
    f: function - The function to integrate
    a: float - The start of the interval
    b: float - The end of the interval

    Returns:
    float - The approximate integral value
    """
    mid = (a + b) / 2
    return (b - a) / 6 * (f(a) + 4 * f(mid) + f(b))

# Example - Approximating the integral of f(x) = x^2 from 0 to 2
"""
Let's use Simpson's Rule to approximate the integral of f(x) = x^2 over the interval [0, 2].
The exact value of the integral is ∫(0 to 2) x^2 dx = (2^3)/3 = 8/3 ≈ 2.6667.
"""

# Define the function
def f(x):
    return x**2

# Define the interval
a = 0
b = 2

# Compute the integral using Simpson's Rule
approximation = simpsons_rule(f, a, b)

# Print the result
print(f"Approximation of ∫(0 to 2) x^2 dx using Simpson's Rule: {approximation:.4f}")

# Visualizing Simpson's Rule


# Generate x values for plotting
x = np.linspace(a, b, 100)
y = f(x)

# Plot the function
plt.plot(x, y, label="f(x) = x^2", color="blue")

# Plot the parabolic approximation
x_parabola = np.linspace(a, b, 100)
y_parabola = f(a) * (x_parabola - (a + b) / 2)**2 + f((a + b) / 2)
plt.fill_between(x_parabola, y_parabola, color="lightblue", alpha=0.5, label="Parabolic Approximation")


plt.scatter([a, (a + b) / 2, b], [f(a), f((a + b) / 2), f(b)], color="red", label="Key Points")


plt.title("Visualization of Simpson's Rule")
plt.xlabel("x")
plt.ylabel("f(x)")
plt.legend()
plt.grid()
plt.show()

```
