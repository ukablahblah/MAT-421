```python
# Importing necessary libraries
import numpy as np
import matplotlib.pyplot as plt

# Root Finding Methods: Bisection Method and Newton-Raphson Method

# Problem Statement
# We aim to find the root of the function f(x) = x^3 - x - 2
# A root is a value of x where f(x) = 0.

# Define the function and its derivative
def f(x):
    return x**3 - x - 2

def f_prime(x):
    return 3*x**2 - 1

# Plot the function to visualize the root
x = np.linspace(-2, 3, 1000)
y = f(x)

plt.figure(figsize=(8, 6))
plt.plot(x, y, label="f(x) = x^3 - x - 2")
plt.axhline(0, color='red', linestyle='--', label="y = 0")
plt.title("Visualization of the Function f(x)")
plt.xlabel("x")
plt.ylabel("f(x)")
plt.legend()
plt.grid()
plt.show()
```

```python

# Tolerance
# Tolerance is the acceptable error in the root value.
# We will use a tolerance of 1e-6 (0.000001) for both methods.

tolerance = 1e-6

```

```python
def f(x):
    return x**3 - 4*x - 9  # Example function

def bisection_method(f, a, b, tol=1e-5, max_iter=100):
    if f(a) * f(b) >= 0:
        raise ValueError("Bisection method requires f(a) and f(b) to have opposite signs.")

    iter_count = 0
    while (b - a) / 2 > tol and iter_count < max_iter:
        c = (a + b) / 2
        if f(c) == 0:
            return c
        elif f(a) * f(c) < 0:
            b = c
        else:
            a = c
        iter_count += 1

    return (a + b) / 2

# Parameters
a, b = 2, 3  # Interval for Bisection Method
tol = 1e-5

# Compute root
root_bisection = bisection_method(f, a, b, tol)

# Print result
print(f"Root found using Bisection Method: {root_bisection:.5f}")
```

```python
def newton_raphson(f, df, x0, tol=1e-5, max_iter=100):
    x = x0
    iter_count = 0
    while abs(f(x)) > tol and iter_count < max_iter:
        x = x - f(x) / df(x)
        iter_count += 1

    return x

def df(x):
    return 3*x**2 - 4  # Derivative of the function

# Initial guess for Newton-Raphson Method
x0 = 2.5

# Compute root
root_newton = newton_raphson(f, df, x0, tol)

# Print result
print(f"Root found using Newton-Raphson Method: {root_newton:.5f}")
```
